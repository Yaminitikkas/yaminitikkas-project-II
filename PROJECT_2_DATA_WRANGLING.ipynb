Z Score

```python
#Find the limit of 'temp' column
upper_limit = df4['temp'].mean() + 3 * df4['temp'].std()
lower_limit = df4['temp'].mean() - 3 * df4['temp'].std() # Subtracting for lower limit

print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
# find the outliers

df4.loc[(df4['temp'] > upper_limit) | (df4['temp'] < lower_limit)]
```

```python

new_df4=df4.loc[(df4['casual'] < upper_limit) | (df4['casual'] > lower_limit)]
print('new data:',len(new_df4))
```

```python
# Print the new shape of the DataFrame

print("New Shape: ", df4.shape)
```

```python
sns.boxplot(new_df4['temp'])
```

```python
# IQR Method for a particular columns

q1=df4['temp'].quantile(.25)
q3=df4['temp'].quantile(.75)
iqr=q3-q1
```

```python
q1, q3, iqr
```

```python
upperlimit=q3+(1.5*iqr)
lower_limit=q1-(1.5*iqr)
upper_limit, lower_limit
```
